```python
import pandas as pd

DATASET_PATH = '../risultati/codeonly-t06.csv'
df = pd.read_csv(DATASET_PATH)
```

```python
NONE                = 0;    NONE_KEY='none'
COMPILE_SUCCESSES   = 1;    COMPILE_KEY='c_compile'
FUNCTIONAL_SUCCESS  = 2;    FUNCTIONAL_KEY='c_functional'
ALMOST_SUCCESS      = 3;    ALMOST_KEY='c_almost'
SUBSET_SUCCESS      = 4;    SUBSET_KEY='c_functional_subset'
SCORES_KEY = 'scores'
N_KEY = 'n'
```

```python
from utils import pass_at_k

COUNT_ALMOST_AS_FUNC = False
COUNT_SUBSET_AS_FUNC = True

compile_match   = [COMPILE_KEY, ALMOST_KEY, SUBSET_KEY, FUNCTIONAL_KEY]
func_match      = [FUNCTIONAL_KEY]

if COUNT_ALMOST_AS_FUNC:    func_match.append(ALMOST_KEY)
if COUNT_SUBSET_AS_FUNC:    func_match.append(SUBSET_KEY)
```

```python
from typing import List


def sum_over_columns(x: dict, columns: List[str]) -> int:
    return sum([int(x[key]) for key in columns])
```

```python
K = 10
COMPILE_K_KEY = f"pass@{K}_compile"
FUNC_K_KEY = f"pass@{K}_func"


df[COMPILE_K_KEY]   = df.apply(lambda x: pass_at_k(x[N_KEY], sum_over_columns(x, compile_match), K), axis=1)
df[FUNC_K_KEY]      = df.apply(lambda x: pass_at_k(x[N_KEY], sum_over_columns(x, func_match), K), axis=1)
```

```python


# Prepare the new DataFrame with the desired structure
out_df = pd.DataFrame()  # This will be the output DataFrame

# Fill the first column with unique 'test' values and the second column (k) with 1, 5, and 10
out_df['test'] = df['test'].repeat(3).reset_index(drop=True)
out_df['k'] = pd.Series([1, 5, 10] * len(df))
out_df['provider'] = df['provider'].repeat(3).reset_index(drop=True)

# Custom function to fill in the pass data for the given 'k' value
def get_pass_data(row, k):
    suffix = f'@{k}'
    compile_key = f'pass{suffix}_compile'
    func_key = f'pass{suffix}_func'
    return pd.Series([row[compile_key], row[func_key], row[compile_key], row[func_key]])

# Create a temporary DataFrame to store 'pass@k' data and then concatenate it with the output DataFrame
compile_checks, functional_checks = [], []
for k in [1, 5, 10]:
    tmp_df = df.apply(lambda row: get_pass_data(row, k), axis=1)
    compile_checks.append(tmp_df.iloc[:,0].values)
    functional_checks.append(tmp_df.iloc[:,1].values)
    
pass_data = pd.DataFrame({
    "Llama-2-codeonly-Compile": compile_checks[0],
    "Llama-2-mix-Compile": compile_checks[1],
    "Llama-2-codeonly-Functional": functional_checks[0],
    "Llama-2-mix-Functional": functional_checks[1]
})
out_df = pd.concat([out_df.reset_index(drop=True), pass_data], axis=1)

```

```python
import numpy as np


# Create a new DataFrame with the desired format
new_data = {
    'test': np.repeat(df['test'].values, 3),
    'k': [1, 5, 10] * len(df),
}

# Prepare the header rows
header = pd.MultiIndex.from_product(
    [['pass@k'], ['Compile check', 'Functional Correctness'], ['Llama-2-codeonly', 'Llama-2-mix']],
    #names=['', 'Criterion', 'Model']
)

# Create the DataFrame to hold the processed data with the headers
out_df = pd.DataFrame(columns=header, data=np.zeros((len(new_data['test']), 4)))

# Fill in the new DataFrame with pass@k values
for k in [1, 5, 10]:
    for index, row in df.iterrows():
        compile_col = f'pass@{k}_compile'
        func_col = f'pass@{k}_func'
        
        # Set NaN values to 0
        compile_val = (row.get(compile_col, 0) or 0)*100
        func_val = (row.get(func_col, 0) or 0)*100
        
        # Find the correct position in the output DataFrame (3 rows per original row - for k = 1, 5, 10)
        out_idx = index * 3 + [1, 5, 10].index(k)
        
        # Update the DataFrame with the values
        out_df.at[out_idx, ('pass@k', 'Compile check', 'Llama-2-codeonly')] = compile_val
        out_df.at[out_idx, ('pass@k', 'Compile check', 'Llama-2-mix')] = compile_val
        out_df.at[out_idx, ('pass@k', 'Functional Correctness', 'Llama-2-codeonly')] = func_val
        out_df.at[out_idx, ('pass@k', 'Functional Correctness', 'Llama-2-mix')] = func_val

# Insert the 'test' and 'k' columns as the first level of the MultiIndex
out_df.insert(0, 'k', new_data['k'])
out_df.insert(0, 'test', new_data['test'])

out_df['provider'] = df['provider'].repeat(3).reset_index(drop=True)
out_df = out_df.round(1)
```

```python
out_df.to_csv('../risultati/codeonly_transformed_data.csv', index=False)
```
