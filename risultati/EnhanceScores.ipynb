```python
import pandas as pd

DATASET_PATH = './mix-t06.csv'
df = pd.read_csv(DATASET_PATH)
```

```python
NONE                = 0;    NONE_KEY='none'
COMPILE_SUCCESSES   = 1;    COMPILE_KEY='c_compile'
FUNCTIONAL_SUCCESS  = 2;    FUNCTIONAL_KEY='c_functional'
ALMOST_SUCCESS      = 3;    ALMOST_KEY='c_almost'
SUBSET_SUCCESS      = 4;    SUBSET_KEY='c_functional_subset'
SCORES_KEY = 'scores'
N_KEY = 'n'
```

```python
values  = [NONE, COMPILE_SUCCESSES, FUNCTIONAL_SUCCESS, ALMOST_SUCCESS, SUBSET_SUCCESS]
keys    = [NONE_KEY, COMPILE_KEY, FUNCTIONAL_KEY, ALMOST_KEY, SUBSET_KEY]

for k,v in zip(keys, values):
    df[k] = df[SCORES_KEY].str.count(str(v))
```

```python
df[N_KEY] = df[SCORES_KEY].str.len()
```
